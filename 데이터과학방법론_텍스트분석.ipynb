```python
import pandas as pd

# 데이터를 pandas DataFrame으로 불러옴
review_data = pd.read_excel("review_20230725_091330.xlsx")
```

SQL 환경설정
* SQLite 이용

```python
import sqlite3

# 1. SQLite3를 사용하여 데이터베이스 파일 생성
conn = sqlite3.connect("reviews.db")

# 2. 해당 데이터베이스에 테이블을 생성하고 데이터를 입력
review_data.to_sql('product_reviews', conn, if_exists='replace', index=False)

# 3. 데이터를 조회하여 확인
query = "SELECT * FROM product_reviews LIMIT 5"
result = pd.read_sql(query, conn)
result.head()
```

기존 sql과 다른점

```python
query = """
  SELECT 상품명, COUNT(리뷰상세내용) AS 리뷰개수
  FROM product_reviews
  GROUP BY 상품명
  ORDER BY 리뷰개수 DESC
"""
result = pd.read_sql(query, conn)
result
```

쥬시만 뽑아보면

```python
query = """
  SELECT 상품명, COUNT(리뷰상세내용) AS 리뷰개수
  FROM product_reviews
  WHERE 상품명 LIKE '%쥬시%'
"""
##왼/오에 뭐가있건 쥬시 있으면 가져와라
result = pd.read_sql(query, conn)
result
```

추가분석
* 조건문 집계

```python
query = """
  select
    case when 리뷰상세내용 like '%쥬시%' then '쥬시'
         when 리뷰상세내용 like '%아메리카노%' then '아메리카노'
         when 리뷰상세내용 like '%에스프레소%' then '에스프레소'
         when 리뷰상세내용 like '%드립백%' then '드립백'
         when 리뷰상세내용 like '%캡슐%' then '캡슐'
         else '기타'
         end as 커피_종류, avg(구매자평점) as 평균_평점
  from product_reviews
  group by 커피_종류
  order by 평균_평점 desc
"""
#case when 에서 한 구문에 두개 이상의 해당단어가 있는지는 확인해봐야함 (즉, 중첩주의)
result = pd.read_sql(query, conn)
result
```

일별 맛 평점 집계

```python
query = """
  SELECT SUBSTR(리뷰등록일, 1, 10) AS 날짜,
      CASE
          WHEN 리뷰상세내용 LIKE '%쥬시%' THEN '쥬시'
          WHEN 리뷰상세내용 LIKE '%아메리카노%' THEN '아메리카노'
          WHEN 리뷰상세내용 LIKE '%에스프레소%' THEN '에스프레소'
          WHEN 리뷰상세내용 LIKE '%드립백%' THEN '드립백'
          WHEN 리뷰상세내용 LIKE '%캡슐%' THEN '캡슐'
          ELSE '기타'
      END AS 커피맛,
      AVG(구매자평점) AS 평균평점
  FROM product_reviews
  GROUP BY SUBSTR(리뷰등록일, 1, 10), 커피맛;
"""
result = pd.read_sql(query, conn)
result
```

시간별 맛 평점 집계

```python
query = """
  SELECT SUBSTR(리뷰등록일, 16, 2) AS 시간,
      CASE
          WHEN 리뷰상세내용 LIKE '%쥬시%' THEN '쥬시'
          WHEN 리뷰상세내용 LIKE '%아메리카노%' THEN '아메리카노'
          WHEN 리뷰상세내용 LIKE '%에스프레소%' THEN '에스프레소'
          WHEN 리뷰상세내용 LIKE '%드립백%' THEN '드립백'
          WHEN 리뷰상세내용 LIKE '%캡슐%' THEN '캡슐'
          ELSE '기타'
      END AS 커피맛,
      AVG(구매자평점) AS 평균평점
  FROM product_reviews
  GROUP BY SUBSTR(리뷰등록일, 16, 2), 커피맛
  order by SUBSTR(리뷰등록일, 16, 2);
"""
result = pd.read_sql(query, conn)
result
```

그럼 캡슐에서, 리뷰4점 이하는 리뷰를 어떻게 썼을까?

```python
query = """
  SELECT 상품명, 구매자평점, 리뷰상세내용
  FROM product_reviews
  where 구매자평점 <= 4
  and 리뷰상세내용 like '%캡슐%'
"""
result = pd.read_sql(query, conn)
result
```

아메리카노 중 가장 오래된 리뷰 찾기

```python
query = """
  SELECT 리뷰상세내용, SUBSTR(리뷰등록일, 1, 10) as 일자
  FROM product_reviews
  WHERE 리뷰상세내용 LIKE '%아메리카노%'
  ORDER BY SUBSTR(리뷰등록일, 1, 10)
  LIMIT 1;
"""
result = pd.read_sql(query, conn)
result
```
